```python
import os
import csv
import json

def getFile():
    data = []
    filePath = './'
    MAX_DEPTH = 2
    for root, dirs, files in os.walk(filePath, topdown=True):
        for name in files:
            if "parser" in name:
                path = os.path.join(root, name)
                parts = path.split('/')
                dirs = "/".join(parts[ : -1])
                data.append([path, dirs, parts[-1]])
    return data

def classifiction():
    data = getFile()
    
    for line in data:
        readPath, savePath = line
        failCount = 1
        
        with open(readPath, newline='') as f:
            reader = csv.reader(f)
            for row in reader:
                if "failed" in row[1]:
                    error = json.loads(row[1])
                    if 'errorMsg' not in error:
                        toSave = savePath + "/calcite_parser_fail/"
                        if not os.path.exists(toSave):
                            os.makedirs(toSave)
                        fileName = "parser_fail_" + str(failCount) + ".txt"
                        failCount += 1
                        fail_file = open(toSave + fileName, "w")
                        fail_file.write(row[0])
                        fail_file.close()
        
```
